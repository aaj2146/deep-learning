```python
a = [ line for line in open('terminal_feedback.txt') if 'loss' in line]
```

```python
b = [ line for line in open('terminal_feedback.txt') if 'accuracy' in line]
```

```python
L_loss = []
for i in range(len(a)):
    L_loss.append(float(a[i].split('=')[2].split(' ')[1]))
```

```python
L_acc = []
for i in range(len(b)):
    L_acc.append(float(b[i].split('accuracy = ')[1].split(',')[0]))
```

```python
x = range(0,95000,100)
```

```python
L_x = []
for i in x: 
    L_x.append(i+100)
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(16,10))
plt.xlabel('step', fontsize=18)
plt.ylabel('loss', fontsize=18)
plt.plot(L_x,L_loss)
plt.show()
```

```python
x_acc = range(1,95,1)
L_x_acc = []
for i in x_acc:
    L_x_acc.append(i)
plt.figure(figsize=(16,10))
plt.xlabel('epoch', fontsize=18)
plt.ylabel('accuracy', fontsize=18)
plt.plot(L_x_acc,L_acc)
plt.show()
```
